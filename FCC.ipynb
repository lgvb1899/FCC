```
df = pd.read_csv('openocean.csv')
longs = []
lats = []
for i in df.index:
  longs += [df.loc[i]['Longitude']]
  lats += [df.loc[i]['Latitude']]
print(longs)
print(lats) 

fig = plt.figure()

ax = plt.axes(projection=ccrs.PlateCarree())
ax.stock_img()

ax.set_extent([-180, 180, -90, 90], crs=ccrs.PlateCarree())

for i in df.index:
  ax.scatter(longs[i], lats[i], color='blue', transform=ccrs.PlateCarree())

ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.COASTLINE)

plt.show()
```

```
df2 = pd.read_csv('bouys.csv')
longs2 = []
lats2 = []
for i in df2.index:
  longs2 += [df2.loc[i]['Longitude']]
  lats2 += [df2.loc[i]['Latitude']]
print(longs2)
print(lats2)

fig = plt.figure()

ax = plt.axes(projection=ccrs.PlateCarree())
ax.stock_img()

ax.set_extent([-180, 0, 0, 90], crs=ccrs.PlateCarree())

for i in df2.index:
  ax.scatter(longs2[i], lats2[i], color='blue', transform=ccrs.PlateCarree())

ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.COASTLINE)

plt.show()
```

```
df3 = pd.read_csv('reefs.csv')
longs3 = []
lats3 = []
for i in df3.index:
  longs3 += [df3.loc[i]['Longitude']]
  lats3 += [df3.loc[i]['Latitude']]
print(longs3)
print(lats3)

fig = plt.figure()

ax = plt.axes(projection=ccrs.PlateCarree())
ax.stock_img()

ax.set_extent([-180, 180, -90, 90], crs=ccrs.PlateCarree())

for i in df3.index:
  ax.scatter(longs3[i], lats3[i], color='blue', transform=ccrs.PlateCarree())

ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.COASTLINE)

plt.show()
```

```
fig = plt.figure()

ax = plt.axes(projection=ccrs.PlateCarree())
ax.stock_img()

ax.set_extent([-180, 180, -90, 90], crs=ccrs.PlateCarree())

total_longs = longs + longs2 + longs3
length = len(total_longs)
total_lats = lats + lats2 + lats3
print(total_longs)
print(length)
print(total_lats)

for i in range(length):
  ax.scatter(total_longs[i], total_lats[i], color='blue', transform=ccrs.PlateCarree())

ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.COASTLINE)

plt.show()
```
